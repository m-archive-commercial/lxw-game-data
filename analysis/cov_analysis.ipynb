### 五大人格与基本特征相关性分析（Pearson）

```python
def transformCell(v, significances = [0.01, 0.05, 0.1]):
    i = 0
    n = len(significances)
    while i < n and significances[i] < v:
        i += 1
    return f"{v:.3f}&nbsp;{'*'*(n-i)}{'&nbsp;' * (i)}"


dfc = df.corr(method='pearson').iloc[:-5, -5:].applymap(transformCell).sort_index()
dfc.style.set_properties(**{
#     "font-size": "24px",
#     "font-family": "monospace"
})
```
